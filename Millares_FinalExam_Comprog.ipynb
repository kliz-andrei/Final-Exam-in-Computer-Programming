# Shampoo Sales Performance Analysis

## Objective
To analyze the sales performance of a shampoo company over a period of time using Python's data analysis and visualization tools.

## Dataset
The dataset contains information about the monthly sales of shampoo. Each entry consists of the month and the corresponding sales amount.

## Steps
1. Load the dataset.
2. Perform data cleaning and preprocessing.
3. Analyze the sales performance.
4. Visualize the sales trends.

# Import necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load the dataset

```python
sales_data = pd.read_csv('C:\\Users\\milla\\OneDrive\\Documents\\Schoolwork\\1st year\\Comprog\\shampoo_sales.csv')
```

# Display the first few rows of the dataset

```python
print("Dataset Preview:")
print(sales_data.head())
```

# Sort the 'Month'

```python
sales_data = sales_data.sort_values(by='Month')
sales_data.set_index('Month', inplace=True)
```

# Calculate the Statistics

```python
total_sales = sales_data['Sales'].sum()
average_sales = sales_data['Sales'].mean()
max_sales = sales_data['Sales'].max()
min_sales = sales_data['Sales'].min()
```

# Print the Output

```python
print("\nDisplay 10 random data of the Shampoo Sales Dataset")
print(sales_data.sample(10))
print(f"Total Sales: {total_sales}")
print(f"Average Sales: {average_sales}")
print(f"Maximum Sales: {max_sales}")
print(f"Minimum Sales: {min_sales}")
```

# Plot the sales data

```python
plt.figure(figsize=(10, 6))
plt.plot(sales_data.index, sales_data['Sales'], marker='o', linestyle='-', color='b')
plt.title('Monthly Shampoo Sales')
plt.xlabel('Month')
plt.ylabel('Sales Amount')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels by 45 degrees
plt.grid(True)
plt.savefig('sales_trends.png')  # Save the plot as an image file
plt.show()
```
